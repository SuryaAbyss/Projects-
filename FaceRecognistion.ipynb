```python
import cv2
import mediapipe as mp
import numpy as np
import csv
from datetime import datetime
from google.colab.patches import cv2_imshow  # For showing images in Colab

# Initialize Mediapipe Face Detection and Face Mesh
mp_face_detection = mp.solutions.face_detection
mp_face_mesh = mp.solutions.face_mesh
face_detection = mp_face_detection.FaceDetection(min_detection_confidence=0.5)
face_mesh = mp_face_mesh.FaceMesh()

# Function to extract unique face landmarks as features
def get_face_encoding(image):
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    results = face_mesh.process(image_rgb)

    if results.multi_face_landmarks:
        landmarks = results.multi_face_landmarks[0].landmark  # Take first detected face
        encoding = np.array([[lm.x, lm.y, lm.z] for lm in landmarks]).flatten()  # Convert to 1D array
        return encoding
    return None

# Load images and extract encodings
known_face_encodings = []
known_face_names = ["Satya Nadella", "Jeff Bezos", "Elon Musk", "Kajal Aggarwal"]
image_files = ["satya_nadella.jpg", "jeff_bezos.jpg", "elon_musk.jpg", "kajal_aggarwal.jpg"]

for image_file in image_files:
    image = cv2.imread(image_file)
    encoding = get_face_encoding(image)
    if encoding is not None:
        known_face_encodings.append(encoding)

# Load a test image for recognition
test_image = cv2.imread("test_image.jpg")  # Upload a test image with a face
test_encoding = get_face_encoding(test_image)

# Recognize faces
face_names = []
if test_encoding is not None:
    distances = [np.linalg.norm(test_encoding - known_encoding) for known_encoding in known_face_encodings]
    best_match_index = np.argmin(distances)
    if distances[best_match_index] < 0.5:  # Adjust threshold if needed
        face_names.append(known_face_names[best_match_index])
    else:
        face_names.append("Unknown")

# Print recognized names
print("Recognized Faces:", face_names)

# Save attendance to CSV
now = datetime.now()
current_date = now.strftime("%Y-%m-%d")

with open(f"{current_date}.csv", "w+", newline="") as f:
    lnwriter = csv.writer(f)
    lnwriter.writerow(["Name", "Time"])
    for name in face_names:
        current_time = now.strftime("%H:%M:%S")
        lnwriter.writerow([name, current_time])
        print(f"{name} marked present at {current_time}")

# Show the test image
cv2_imshow(test_image)
```

```python
import cv2
import mediapipe as mp
import numpy as np
import os
import csv
from datetime import datetime, timedelta
from collections import defaultdict

# ================= Configuration =================
IMAGE_DIR = r"C:\Python\images"
CSV_DIR = os.path.expanduser("~\\Documents\\Attendance")
# ==================================================

# Initialize MediaPipe components with optimized parameters
mp_face_detection = mp.solutions.face_detection
mp_face_mesh = mp.solutions.face_mesh

face_detection = mp_face_detection.FaceDetection(
    min_detection_confidence=0.7,
    model_selection=1  # Use accurate model
)

face_mesh = mp_face_mesh.FaceMesh(
    static_image_mode=True,
    max_num_faces=1,
    min_detection_confidence=0.6,
    min_tracking_confidence=0.5
)

def get_face_encoding(image):
    """Robust face encoding with enhanced preprocessing and alignment"""
    # Convert to RGB and detect faces
    rgb_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    results = face_detection.process(rgb_image)

    if not results.detections:
        return None

    # Select largest face with boundary checks
    ih, iw = image.shape[:2]
    face_areas = []
    for detection in results.detections:
        bbox = detection.location_data.relative_bounding_box
        face_areas.append(bbox.width * bbox.height)

    main_face_idx = np.argmax(face_areas)
    bbox = results.detections[main_face_idx].location_data.relative_bounding_box

    # Calculate expanded face region
    x = int(bbox.xmin * iw)
    y = int(bbox.ymin * ih)
    w = int(bbox.width * iw)
    h = int(bbox.height * ih)

    # Expand ROI by 25% with safety checks
    x = max(0, x - int(w * 0.25))
    y = max(0, y - int(h * 0.25))
    w = min(iw - x, int(w * 1.5))
    h = min(ih - y, int(h * 1.5))

    face_roi = image[y:y+h, x:x+w]
    if face_roi.size == 0:
        return None

    # Enhanced preprocessing pipeline
    face_roi = cv2.resize(face_roi, (500, 500))
    gray = cv2.cvtColor(face_roi, cv2.COLOR_BGR2GRAY)
    clahe = cv2.createCLAHE(clipLimit=3.0, tileGridSize=(8, 8))
    processed_face = clahe.apply(gray)
    processed_face = cv2.GaussianBlur(processed_face, (3, 3), 0)

    # Get facial landmarks
    results_mesh = face_mesh.process(cv2.cvtColor(processed_face, cv2.COLOR_GRAY2RGB))

    if not results_mesh.multi_face_landmarks:
        return None

    # Use stable facial landmarks for encoding
    LANDMARK_IDS = [33, 133, 159, 263,  # Eyes
                    1, 4, 195, 5,  # Nose
                    61, 291, 78, 306]  # Mouth
    landmarks = results_mesh.multi_face_landmarks[0].landmark

    encoding = np.array([(landmarks[i].x, landmarks[i].y, landmarks[i].z)
                         for i in LANDMARK_IDS]).flatten()
    norm = np.linalg.norm(encoding)
    return encoding / norm if norm != 0 else None

def load_reference_images():
    """Load and validate reference images with multiple detection attempts"""
    known_encodings = []
    known_names = []

    reference_files = {
        "satya_nadella.jpg": "Satya Nadella",
        "jeff_bezos.jpg": "Jeff Bezos",
        "elon_musk.jpg": "Elon Musk",
        "kajal_aggarwal.jpg": "Kajal Aggarwal"
    }

    print("\nLoading reference images:")
    for filename, name in reference_files.items():
        path = os.path.join(IMAGE_DIR, filename)
        if not os.path.exists(path):
            print(f"❌ File missing: {filename}")
            continue

        image = cv2.imread(path)
        if image is None:
            print(f"❌ Corrupted file: {filename}")
            continue

        # Maintain aspect ratio while resizing
        h, w = image.shape[:2]
        scale = 800 / max(h, w)
        image = cv2.resize(image, (int(w*scale), int(h*scale)))

        # Try multiple detection strategies
        for attempt in range(3):
            if attempt == 1:
                image = cv2.rotate(image, cv2.ROTATE_90_CLOCKWISE)
            elif attempt == 2:
                image = cv2.flip(image, 1)

            encoding = get_face_encoding(image)
            if encoding is not None:
                known_encodings.append(encoding)
                known_names.append(name)
                print(f"✅ Loaded: {filename} (attempt {attempt+1})")
                break
        else:
            print(f"⚠️ Face not detected in {filename}")
            print("  Check: Front-facing, good lighting, no obstructions")

    return known_encodings, known_names

# Initialize reference database
known_encodings, known_names = load_reference_images()

if not known_encodings:
    print("\n❌ Critical Error: No valid reference faces found")
    print("Troubleshooting steps:")
    print("1. Verify images are in JPG format")
    print("2. Ensure faces are clearly visible and centered")
    print("3. Check minimum image size (600x600 pixels)")
    print(f"4. Confirm directory exists: {IMAGE_DIR}")
    exit()

# Calculate dynamic thresholds
def calculate_thresholds(encodings):
    distances = []
    for i in range(len(encodings)):
        for j in range(i+1, len(encodings)):
            distances.append(np.linalg.norm(encodings[i] - encodings[j]))
    return {
        'match': np.mean(distances) * 0.7,
        'reject': np.mean(distances) * 1.3
    }

thresholds = calculate_thresholds(known_encodings)
print(f"\n⚙️ Recognition thresholds - Match: {thresholds['match']:.2f}, Reject: {thresholds['reject']:.2f}")

# Initialize video capture
cap = cv2.VideoCapture(0)
attendance_log = defaultdict(lambda: datetime.min)
confidence_scores = defaultdict(float)
CONFIDENCE_DECAY = 0.85
REQUIRED_CONFIDENCE = 0.92

os.makedirs(CSV_DIR, exist_ok=True)

while True:
    success, frame = cap.read()
    if not success:
        break

    frame = cv2.flip(frame, 1)
    current_time = datetime.now()

    # Process frame
    encoding = get_face_encoding(frame)
    if encoding is None:
        cv2.putText(frame, "No face detected", (20, 40),
                   cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 0, 255), 2)
        cv2.imshow('Face Attendance System', frame)
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
        continue

    # Calculate similarities
    similarities = [1 - np.linalg.norm(encoding - known_enc) for known_enc in known_encodings]
    best_match_idx = np.argmax(similarities)
    best_score = similarities[best_match_idx]
    best_name = known_names[best_match_idx]

    # Update confidence scores
    for name in confidence_scores:
        confidence_scores[name] *= CONFIDENCE_DECAY
    confidence_scores[best_name] += best_score

    # Determine recognition status
    recognition_status = "Unknown"
    if confidence_scores[best_name] > REQUIRED_CONFIDENCE:
        if best_score > (1 - thresholds['match']):
            recognition_status = best_name
    elif max(confidence_scores.values(), default=0) < 0.4:
        recognition_status = "Unknown"

    # Handle attendance logging
    if recognition_status != "Unknown":
        if current_time - attendance_log[recognition_status] > timedelta(minutes=5):
            attendance_log[recognition_status] = current_time
            csv_path = os.path.join(CSV_DIR, f"attendance_{current_time.date()}.csv")
            with open(csv_path, 'a', newline='') as f:
                writer = csv.writer(f)
                writer.writerow([recognition_status, current_time.strftime("%Y-%m-%d %H:%M:%S")])
            print(f"✅ Attendance marked: {recognition_status}")
            confidence_scores.clear()

    # Display results
    color = (0, 255, 0) if recognition_status != "Unknown" else (0, 0, 255)
    cv2.putText(frame, f"Status: {recognition_status}", (20, 40),
               cv2.FONT_HERSHEY_SIMPLEX, 0.8, color, 2)
    cv2.putText(frame, f"Confidence: {best_score:.2f}", (20, 80),
               cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 255, 255), 2)

    cv2.imshow('Face Attendance System', frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
print("System shutdown successfully")
```
